```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Iris.csv')
df.head(10)
```

# scatter plot for sepailenghthcm and sepalwidth

```python
sns.scatterplot(x=df['SepalLengthCm'],y=df['SepalWidthCm'])
plt.show()
```

# countplot for species

```python

sns.countplot(x=df['Species'])
plt.show()
```

# scatter plot for sepailenghthcm and sepalwidth with hue of species

```python
sns.scatterplot(x=df['SepalLengthCm'],y=df['SepalWidthCm'],data=df,hue=df['Species'])
plt.show()
```

# Relation plot between PetalLengthCm and PetalWidthCm

```python
sns.relplot(x=df['PetalLengthCm'],y=df['PetalWidthCm'],kind='scatter',data=df,col=df['Species'])
plt.show()
```

# point size

```python
sns.relplot(x=df['PetalLengthCm'],y=df['PetalWidthCm'],kind='scatter',data=df,size=
df['PetalLengthCm'],hue=df['PetalLengthCm'])
plt.show()
```

# style

```python
sns.relplot(x=df['PetalLengthCm'],y=df['PetalWidthCm'],kind='scatter',data=df,size=
df['PetalLengthCm'],hue=df['PetalLengthCm'],style=df['Species'])
plt.show()
```

# line plot botween sepailenghthcm and sepalwidth 

```python
sns.relplot(x=df['SepalLengthCm'],y=df['SepalWidthCm'],data=df,kind='line')
plt.show()
```

# Bar plot for  sepailenghthcm and  species

```python
sns.catplot(y=df['SepalLengthCm'],x=df['Species'],data=df,kind='bar')
plt.show()
```

# Box plot

```python
sns.boxplot(x="Species", y="PetalLengthCm", palette="husl", data=df)
```
